```python
def show_histogram_seaborn(merged_data):
        histogram = sns.displot(merged_data, x="LINK-USD",y="DOT-USD",hue="UNI1-USD",kind="hist",rug=True)
        sns.color_palette("light:#5A9", as_cmap=True)
        histogram.set_axis_labels("LINK-USD","DOT-USD")
        

show_histogram_seaborn(return_grouped_data_frame('High',crypto_list).head(8))
```

```python
def show_kernel_destiny_estimate_seaborn(merged_data):
        sns.kdeplot(data=merged_data, x="UNI1-USD",y="DOT-USD",shade=False,bw_adjust=1, color="green",vertical=False,legend=True,cumulative=True,cbar=True)  

show_kernel_destiny_estimate_seaborn(return_grouped_data_frame('High',crypto_list).head(8))
```
